## matplot lib example workflow

```python
#1 prepare data
x = [1,2,3,4]
y = [11,3,22,55]

#2 setup plot
fig,ax = plt.subplots(figsize=(10,10))

#3 plot data
ax.plot(x,y)

#4 customize
ax.set(title="simple point",
      xlabel="x-axis",
      ylabel="y-axis")

#5 save and show
fig.savefig("./../l1/sample_plot_patric.png")
```

```python
import numpy as np
```

```python
#creating data
x = np.linspace(0,10,100)
x[:10]
```

```python
#plot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#use data to create scatter plot
fix,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
#scatter plot tomorrow
fig,ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# making a plot from dictionary
nut_butter_prices={"Almond Butter": 10,
                  "Peanut Butter" : 8,
                  "Cashew Butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's Nut Buuter Store",
      ylabel="prices ($)")
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),nut_butter_prices.values());
```

```python
#some random data
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

# creating subplots

```python
#type 1

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.hist(x);
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax3.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
ax4.scatter(x,np.exp(x));
```
